**UNEMPLOYMENT ANALYSIS USING PYTHON**

```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#upload the data

data = pd.read_csv('Unemployment in India.csv')
data = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
print(data.head())
```

```python
#dataset missing values

print(data.isnull().sum())
```

```python
#understanding better missing columns

data.columns=['States','Date','	Frequency','Estimated Unemployment Rate','Estimated Employee','Estmated Labour Participation Rate','Region','Longitude','Lattitude']
```

```python
#correlation

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sns.heatmap(data.corr())
plt.show()
```

```python
#data visualization
data.columns=['States','Date','	Frequency','Estimated Unemployment Rate','Estimated Employee','Estmated Labour Participation Rate','Region','Longitude','Lattitude']
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Employee',hue='Region',data = data)
plt.show()
```

```python
#unemployment rate according to different regions of india

plt.figure(figsize=(12,10))
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Unemployment Rate',hue='Region',data = data)
plt.show()
```

```python
#dashboard of analyze the unemployment rate of each indian spirit
unemployment = data[['States','Region','Estimated Unemployment Rate']]
figure=px.sunburst(unemployment ,path=['Region','States'],values = 'Estimated Unemployment Rate',width=700,height = 700,color_continuous_scale= "RdYlGn",title="Unemployment Rate In India")
figure.show()
```
